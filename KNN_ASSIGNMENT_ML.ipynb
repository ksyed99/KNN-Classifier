```python
from scipy import misc #miscellaneous services 
import cv2 #opencv used in image processing
from math import sqrt #sqrt function to be used later 
from numpy import ndarray # ndarray is a multidimensional container of items of the same type and size.
from scipy.spatial.distance import cdist #data manipulation and visualisation
import cv2
from google.colab.patches import cv2_imshow #google colab alternative to cv2.imshow() 


#Read your image here
####### Your code here #######   
my_image = cv2.imread('Fawaddd.jpg') #load my image into the variable
cv2_imshow(my_image) #display my image
############################      


#Resize your image
####### Your code here #######
dimensions=(32,32) #desired dimensions 
resized_image = cv2.resize(my_image, dimensions, interpolation = cv2.INTER_AREA) #resize the image
print('Resized Dimensions : ',resized_image.shape) #display resized dimensions
print('The resized photo: ')
plt.imshow(resized_image) #display the resized image    
cv2.waitKey(0) #so that the picture is displayed for infinite amount of time
cv2.destroyAllWindows() #destroy any windows created
##############################
      
      
#Reshape your image as we reshape the image of dataset
####### Your code here #######    
reshaped_image = np.reshape(resized_image, [3072,1], order="F") #reshape the resized image to (3072,1)
print('Reshaped dimensions: ', reshaped_image.shape) #display the reshaped dimensions
#############################


#Calculate Eucleadian distance between your image and dataset 
####### Your code here #######
def euclidean_distance(row1,row2): #function definition
  distance=0.0 #initializing variable to zero
  for i in range(len(row1)): #loop to traverse through all elements of row
    distance=distance+(row1[i]-row2[i])**2 #part of distance formula. The result from all elements are added
  return sqrt(distance) #distance for 1 row finally calculated
```

```python
dist =np.zeros(50) #initializing array with zeros
for i in range(50): #loop to include all 50 images
  selected_image= images[0:3072,i] #(0-3072) so that distance from each pixel is calculated. For each image, 3072 pixels are added in one array and sent to the euclidean function  
  dist[i]=euclidean_distance(reshaped_image,selected_image) #calculate distance for each image and store in array 

print ("Matrix representing Euclidean distances: \n",dist) #display array containing all distances
print ("\nMinimum Distance = ",min(dist)) #display smallest value in the array
minimumdist=np.argmin(dist) #stores the point at which minimum distance occurs
print("\nIndex of Minimum Distance = ",minimumdist) #display the point at which minimum distance occurs
plt.imshow(im[:,:,:,minimumdist]) #display the image
sortedarray=np.argsort(dist) #create an array with the indexes arranged in ascending order
print("\nArray with indices of distances in ascending order:\n", sortedarray) #display the array
print("\nPhoto with Minimum Distance:\n")
##############################
```

```python
# Write code for 1 NN
#Find min distance
#Fine at which point min value exists

####### Your code here #######
x=sortedarray[0] #store index of smallest distance in x
print("Prediction: You resemble ",label[x][0][0]) #display the name of the person in the corresponding image. From 2d array to 1d array then extract only the value in the array 
print("\nDistance = ",dist[x]) #display the distance
print("Value exists at index = ", x) #display the value(index) at which the distance occurs
plt.imshow(im[:,:,:,x]) #display the image
##############################
```

```python
import statistics
#3 NN
#Write code for 3 NN
#Find 3 min distances
#Find their instances
#hint: Use for Loop

####### Your code here #######
K3NN_labels = [] #array to store the names of the people in the three images 
for i in range(3): #iterate 3 times as k=3

  x=sortedarray[i] #store the index of ith nearest image (i.e. having smallest distance) in x
  K3NN_labels.append(label[x][0][0]) #insert the 1st element in the xth row (name) into the new array  
  print(label[x][0][0]) #display the name of image in the xth row
  print('Distance=', dist[x]) #display distance of the ith nearest image
  print("Value exists at index = ", x) #display the value at which this distance occurs
  plt.imshow(im[:,:,:,x]) #display the image
  plt.show() #added since more than one image has to be displayed
  
 
print(K3NN_labels) #display the array with all three names
try: #block to execute when succesful prediction is made
  print("Prediction: You resemble ",statistics.mode(K3NN_labels)) #display the most frequent name in the array (majority vote)
except : #block to execute when no mode obtained
  print("No single mode found, hence the result is inconclusive.") #let the user know that results are inconclusive
##############################
```
